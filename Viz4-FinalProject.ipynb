```python
import pandas as pd
import matplotlib.pyplot as plt
plt.get_backend
import datetime as dt
import numpy as np

plt.style.use('seaborn-poster')
fig,((ax1,ax2)) = plt.subplots(2, 1, sharex=True, sharey=False)
ocean_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\sea-surface-temp_fig-1.csv', index_col=0, skiprows=6)
ocean_df=ocean_df['Annual anomaly']
ocean_df=ocean_df.to_frame()
ocean_df=ocean_df.ix['1950':'2015']
#ocean_df.index = pd.to_datetime(ocean_df.index)
ax1.plot(ocean_df)
ax1.set_title('Ocean Surface Temperature Anomalies 1950-2015')
ax2.set_title('Occurrences of Extreme Weather in Berks County, PA')

```

```python
columns_to_keep = ['EVENT_TYPE']

tornado_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\storm_data_search_results tornado.csv', index_col=3,\
                         skiprows=0)


tornado_df = tornado_df[columns_to_keep]
tornado_df['TORNADO_COUNT']=1
tornado_df['Year'] = pd.to_datetime(tornado_df.index)
tornado_df['Year'] = tornado_df['Year'].apply(lambda x: int(dt.datetime.strftime(x, '%Y')))
tornado_df=tornado_df.set_index('Year')
tornado_df=tornado_df.groupby([tornado_df.index]).sum()
ax2.scatter(tornado_df.index, tornado_df['TORNADO_COUNT'],label='Tornado Count')

#tornado_df
blizzard_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\storm_data_search_results blizzard.csv', index_col=3, skiprows=0)
blizzard_df = blizzard_df[columns_to_keep]
blizzard_df['BLIZZARD_COUNT']=1
blizzard_df['Year'] = pd.to_datetime(blizzard_df.index)
blizzard_df['Year'] = blizzard_df['Year'].apply(lambda x: int(dt.datetime.strftime(x, '%Y')))
blizzard_df=blizzard_df.set_index('Year')
blizzard_df=blizzard_df.groupby([blizzard_df.index]).sum()
ax2.scatter(blizzard_df.index, blizzard_df['BLIZZARD_COUNT'],label='Blizzard Count')


#blizzard_df
hail_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\storm_data_search_results hail.csv', index_col=3, skiprows=0)
hail_df=hail_df[columns_to_keep]
hail_df['HAIL_COUNT']=1
hail_df['Year'] = pd.to_datetime(hail_df.index)
hail_df['Year'] = hail_df['Year'].apply(lambda x: int(dt.datetime.strftime(x, '%Y')))
hail_df = hail_df.set_index('Year')
hail_df = hail_df.groupby([hail_df.index]).sum()
ax2.scatter(hail_df.index, hail_df['HAIL_COUNT'],label='Hail Count')

#tropical storms
tropical_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\storm_data_search_results tropical storm.csv', index_col=3, skiprows=0)
tropical_df=tropical_df[columns_to_keep]
tropical_df['TROPICAL_STORM_COUNT']=1
tropical_df['Year'] = pd.to_datetime(tropical_df.index)
tropical_df['Year']=tropical_df['Year'].apply(lambda x: int(dt.datetime.strftime(x, '%Y')))
tropical_df = tropical_df.set_index('Year')
tropical_df = tropical_df.groupby([tropical_df.index]).sum()
ax2.scatter(tropical_df.index, tropical_df['TROPICAL_STORM_COUNT'],label='Tropical Storm Count')

#funnel clouds
funnel_df = pd.read_csv('E:\Data Science\Visualizations\Visualization Final Project\cloud.csv')
funnel_df=funnel_df[columns_to_keep]
funnel_df['FUNNEL_CLOUD_COUNT']=1
funnel_df['Year'] = pd.to_datetime(funnel_df.index)
funnel_df['Year'] = funnel_df['Year'].apply(lambda x: int(dt.datetime.strftime(x, '%Y')))
funnel_df = funnel_df.set_index('Year')
funnel_df = funnel_df.groupby([funnel_df.index]).sum()
ax2.scatter(funnel_df.index, funnel_df['FUNNEL_CLOUD_COUNT'],label='Funnel Cloud Count')
ax2.legend(fontsize='medium')
fig.savefig('bricker_vis4_final_berks.jpg')
```
